```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import f1_score
import string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import re

# Load the text data into a pandas dataframe

def preprocess_text(text):
    # Convert all text to lowercase
    text = text.lower()
    
    # Remove numbers and special characters
    text = re.sub(r'\d+', '', text)
    text = re.sub('wafi ', 'wifi ', text)
    text = re.sub('ena ', 'i ', text)
    text = re.sub('kivach ', 'how ', text)
    text = re.sub('top nat ', 'topnet ', text)
    text = re.sub("bon d'achat ", 'shopping voucher ', text)
    text = re.sub('bon d achat ', 'shopping voucher ', text)
    text = re.sub('my tt ', 'mytt ', text)
    text = re.sub('tt ', 'tunisie telecome ', text)
    
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    # Tokenize the text
    tokens = text.split()
    
    # Remove stop words
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if not word in stop_words]
    
    
    
    # Convert the list of tokens back to a string
    text = ' '.join(tokens)
    
    return text

# Apply the clean_text function to the Comments column
# Instantiate the VADER sentiment analyzer
analyzer = SentimentIntensityAnalyzer()
text = preprocess_text(text)
new_words = {
            'improve': 0.7,
            'connect': 0.7,
            'cut': -0.7,
            'waiting': -0.7
    
            }

# Update VADER's lexicon with the new words and scores
analyzer.lexicon.update(new_words)
threshold = 0.0

# Create empty lists to store the predicted sentiment scores and labels
pos_scores = []
neg_scores = []
neu_scores = []
compound_scores = []
Predicted_labels = []

# Iterate over the text data in the dataset and predict the sentiment score and label for each text
comments = df['Comments']  # Get the list of comments
for comment in comments:
    comment_list = eval(comment)
    if len(comment_list) >= 2:
        for p in comment_list:
            text = p[1]
            text = preprocess_text(text)
            score = analyzer.polarity_scores(text)
            pos_scores.append(score["pos"])
            neg_scores.append(score["neg"])
            neu_scores.append(score["neu"])
            compound_scores.append(score["compound"])
            if any(word in text for word in negative_words):
                label = "negative"
            elif score["compound"] >= 0.1:
                label = "positive"
            elif score["compound"] <= -0.3:
                label = "negative"
            else:
                label = "neutral"

            # Add the label to the comment array
            p.append(label)
            Predicted_labels.append(label)
            print(text)
            print(label)
            
    df.loc[df['Comments'] == comment, 'Comments'] = str(comment_list)

# Calculate the overall accuracy of the VADER model
accuracy = accuracy_score(true_labels, Predicted_labels)
report = classification_report(true_labels, Predicted_labels)

print(report)
from sklearn.metrics import precision_score

precision = precision_score(true_labels, Predicted_labels, pos_label='positive')
f1 = f1_score(true_labels, Predicted_labels, average='weighted')
print(f'precision: {precision:.3f}')
print(f'Accuracy: {accuracy:.3f}')
print(f'F1 score: {f1:.3f}')
df.to_csv("data_with_scores.csv", index=False)
```

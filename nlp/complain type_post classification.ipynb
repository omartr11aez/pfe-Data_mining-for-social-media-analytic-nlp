```python
nlp = spacy.load('en_core_web_sm')
# Lemmatize and remove stop words from the complaint patterns
lemmatized_patterns = []
for pattern in complain_patterns:
    lemmatized_pattern = []
    for token_dict in pattern:
        token_dict_processed = {}
        for key, value in token_dict.items():
            if key == "LOWER":
                token_dict_processed[key] = value.lower()
            else:
                token_dict_processed[key] = value
        lemmatized_pattern.append(token_dict_processed)
    lemmatized_patterns.append(lemmatized_pattern)

# Add the lemmatized patterns to the matcher
matcher = Matcher(nlp.vocab)
for i, pattern in enumerate(lemmatized_patterns):
    matcher.add("Complaint", [pattern])

# Define a function to check if a text contains a complaint
def contains_complaint(text):
    doc = nlp(text)
    matches = matcher(doc)
    return len(matches) > 0

df = pd.read_csv("news_post_with_topics.csv")

comments = df['Comments'].tolist()  # Get the list of comments as a Python list
for i, comment in enumerate(comments):
    comment_list = eval(comment)
    if len(comment_list) >= 2:
        for p in comment_list:
            text = p[1]
            # Lemmatize and remove stop words from the comment text
            doc = nlp(text)
            lemmatized_text = " ".join([token.lemma_ for token in doc if not token.is_stop])
            if contains_complaint(lemmatized_text):
                label = "a complaint"
            else:
                label = "not a complaint"
            print(f"Text: {text}\nLabel: {label}\n")
            p.append(label)
    comments[i] = str(comment_list)  # Update the comment in the comments list
    df.loc[df['Comments'] == comment, 'Comments'] = str(comment_list)

df.to_csv("data_with_complaint.csv", index=False)
```

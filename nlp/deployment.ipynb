```python
import numpy as np 
import pandas as pd
```

```python
data = pd.read_csv('data_with_complain.csv', encoding='utf-8')
data['Post']
print("the numbers of the post with null",len(data['Post'].isnull()))
# Drop rows with null values
data = data[data['Post'].notnull()]

# Drop columns with null values
data.dropna(axis='columns')
print("the numbers of the post after removing the null",len(data['Post']))
data.head()
```

```python
data1 = pd.read_csv('react1.csv', encoding='utf-8')
data1['PostId']
print("the numbers of the post with null",len(data1['PostId'].isnull()))
# Drop rows with null values
data1 = data1[data1['PostId'].notnull()]
# Drop columns with null values
data1.dropna(axis='columns')
print("the numbers of the post after removing the null",len(data1['PostId']))
data1.head()
```

```python
data['PostId'].head(5)
import re
pattern = r'&eav.*'

# apply pattern to each value in the 'PostId' column using the apply() method
data['PostId'] = data['PostId'].apply(lambda x: re.sub(pattern, '', x))
data['PostId'].head(5)
```

```python
data1['PostId'].head(5)
import re
pattern = r'&eav.*'

data1['PostId'] = data1['PostId'].apply(lambda x: re.sub(pattern, '', x))
data1['PostId'].head(5)
```

```python
merged_df = pd.merge(data, data1, on='PostId', how='inner')
```

```python
merged_df.head(5)
```

```python
print("the numbers of the post after removing the null",len(merged_df))
```

```python
import re

merged_df['Like'] = merged_df['Reaction'].str.extract(r'(\d+) people reacted with Like')
merged_df['Love'] = merged_df['Reaction'].str.extract(r'(\d+) people reacted with Love')
merged_df['Care'] = merged_df['Reaction'].str.extract(r'(\d+) people reacted with Care')
merged_df['Haha'] = merged_df['Reaction'].str.extract(r'(\d+) people reacted with Haha')
merged_df['Wow'] = merged_df['Reaction'].str.extract(r'(\d+) people reacted with Wow')
merged_df['Angry'] = merged_df['Reaction'].str.extract(r'(\d+) people reacted with Angry')
merged_df['Sad'] = merged_df['Reaction'].str.extract(r'(\d+) person reacted with Sad')

merged_df['Total_Reactions'] = merged_df[['Like', 'Love', 'Care', 'Haha', 'Wow', 'Angry', 'Sad']].astype(float).sum(axis=1)
```

```python
merged_df.head(5)
```

```python
merged_df[['Like', 'Love', 'Care', 'Haha', 'Wow', 'Angry', 'Sad']] = merged_df[['Like', 'Love', 'Care', 'Haha', 'Wow', 'Angry', 'Sad']].fillna(0)
```

```python
merged_df.head(5)
```

```python
merged_df.drop('Reaction', axis=1, inplace=True)
```

```python
import re

pattern1 = r'&eav.*'
pattern2 = r'\?.*'
import re
import ast

merged_df['Name_reaction'] = merged_df['Name_reaction'].apply(lambda x: ast.literal_eval(x))
merged_df['Name_reaction'] = merged_df['Name_reaction'].apply(lambda x: [re.sub(r'[?&]eav.*', '', id) for id in x])
merged_df['Name_reaction'] = merged_df['Name_reaction'].apply(lambda x: str(x))
```

```python
import re
import ast

comments = merged_df['Comments']  # Get the list of comments

for i, comment in enumerate(comments):
    comment_list = ast.literal_eval(comment)
    if len(comment_list) >= 2:
        for p in comment_list:
            text = p[2]
            updated_id = re.sub(r'[?&]eav.*', '', text)  # remove everything after '&eav' or '?eav'
            p[2] = updated_id
        merged_df.loc[i, 'Comments'] = str(comment_list)
```

```python
from collections import Counter

id_freq_dict = {}

# Loop through all the rows in the DataFrame
for i in range(len(merged_df)):
    reaction_list = ast.literal_eval(merged_df.loc[i, 'Name_reaction'])

    # Count the frequency of each id
    id_freq = Counter(reaction_list)

    # Add the frequency of each id to the dictionary
    for id, freq in id_freq.items():
        if id not in id_freq_dict:
            id_freq_dict[id] = freq
        else:
            id_freq_dict[id] += freq
sorted_freq_dict = dict(sorted(id_freq_dict.items(), key=lambda item: item[1], reverse=True))
# Print the frequency of each id across all the Name_reaction columns
for id, freq in sorted_freq_dict.items():
    print(f'{id}: {freq}')
```

```python
from collections import Counter

# Count the frequency of each ID in the 'Name_reaction' column
id_freq_dict = {}
for i in range(len(merged_df)):
    reaction_list = ast.literal_eval(merged_df.loc[i, 'Name_reaction'])
    id_freq = Counter(reaction_list)
    for id, freq in id_freq.items():
        if id not in id_freq_dict:
            id_freq_dict[id] = freq
        else:
            id_freq_dict[id] += freq

# Count the frequency of each text in the 'Comments' column
text_freq_dict = {}
for comments in merged_df['Comments']:
    for comment in ast.literal_eval(comments):
        if len(comment) >= 3:
            text = comment[2]
            if text not in text_freq_dict:
                text_freq_dict[text] = 1
            else:
                text_freq_dict[text] += 1

# Add up the frequencies from both dictionaries
total_freq_dict = {}
for id, freq in id_freq_dict.items():
    total_freq_dict[id] = freq
for text, freq in text_freq_dict.items():
    if text not in total_freq_dict:
        total_freq_dict[text] = freq
    else:
        total_freq_dict[text] += freq

# Sort the result in descending order
sorted_freq = sorted(total_freq_dict.items(), key=lambda x: x[1], reverse=True)

# Remove the first item
sorted_freq = sorted_freq[1:]

# Print the result
for item, freq in sorted_freq:
    print(f'{item}: {freq}')
```
